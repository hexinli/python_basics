## 1.map(function, iterable, ...)

![本地图片](./images/222.PNG)

```python
from parso.python.tree import Lambda

# 单个可迭代对象
nums = [1, 2, 3, 4]
squared = map(lambda x: x**2, nums)
print(squared)
list(squared)
```

```python
# 多个可迭代对象（并行处理）
b = [10, 20, 30, 40]   # 第4个元素被忽略
sums = map(lambda x, y: x + y, a, b)
list(sums)
```

```python
# 使用已有函数（更简洁）
a=list(map(str, [1, 2, 3]))
b=list(map(int, ['10', '20', '30']))
print(a)
print(b)
```

#### 重要特性
* 惰性求值：map 返回的迭代器只能遍历一次。
* 长度截断：当有多个可迭代对象时，结果长度由最短的输入决定，多余元素被静默丢弃（不报错）。
* 不修改原对象：map 产生新迭代器，原可迭代对象不变。
* 函数为 None 时：会抛出 TypeError（与 filter 不同）。

#### 适用场景
* 对序列每个元素进行相同的、无副作用的转换。
* 与已经定义好的单参数或多参数函数配合，避免显式循环。
* 在函数式编程流水线中作为中间步骤。

## 2. filter(function, iterable)

![本地图片](./images/333.PNG)

```python
# 基本筛选
nums = [1, 2, 3, 4, 5, 6]
result = filter(lambda num: num % 2 == 0, nums)
print(list(result))
```

```python
# function=None 时过滤假值
mixed = [0, 1, '', 'hello', None, [], [1, 2]]
result = filter(None, mixed)
print(list(result))
```

```python
# 与 map 组合：先过滤再转换
nums = [1, 2, 3, 4, 5, 6]
result = map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, nums))
print(list(result))
```

#### 与 map 的区别
* map：转换每个元素，输出长度等于输入长度（多迭代器时取最短）。
* filter：根据条件挑选元素，输出长度** ≤ **输入长度。

#### 注意点
*  在 Python 2 中 filter 返回列表，Python 3 返回迭代器。
* 如果 function 返回非布尔值，会被隐式转换为布尔值（遵循 Python 的真值测试规则）。
* 对于同时需要转换和筛选的操作，列表推导/生成器表达式通常更清晰：
(f(x) for x in seq if cond) 等价于 map(f, filter(cond, seq))。

## 3. zip(*iterables)

![本地图片](./images/444.PNG)

```python
# 基本打包
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
list(zip(names, ages))
[('Alice', 25), ('Bob', 30), ('Charlie', 35)]
```

```python
# 长度截断
a = [1, 2, 3, 4]
b = ['a', 'b']
list(zip(a, b))
[(1, 'a'), (2, 'b')]   # 多余元素被忽略
```

```python
# 解压（unzip）
pairs = [('a', 1), ('b', 2), ('c', 3)]
letters, numbers = zip(*pairs)
print(letters)
print(numbers)
```

```python
# 与 map 结合
a = [1, 2, 3]
b = [10, 20, 30]
# 方法1：直接使用多参数 map
res1 = list(map(lambda x, y: x + y, a, b))
print(res1)

# 方法2：先 zip 再 map（展示两者关系）
res2 = list(map(lambda pair: pair[0] + pair[1], zip(a, b)))
print(res2)
```

### 重要特性
* 惰性求值：返回的迭代器只能遍历一次。
* 长度截断：以最短的输入为准，多余部分丢弃（不报错）。
* 可接受零个参数：zip() 返回空迭代器。
* 可接受一个参数：zip([1,2,3]) 产生 (1,), (2,), (3,)。

### 适用场景
* 并行遍历多个序列。
* 将多个序列按位置合并为元组列表。
* 配合 dict() 快速构建字典：dict(zip(keys, values))。
* 实现矩阵转置（配合 * 解包）。

```python
# 参考示例：函数式流水线
# 需求：计算 1~10 中偶数的平方，然后与字母 'a','b','c','d','e' 组合
numbers = range(1, 11)
letters = ['a', 'b', 'c', 'd', 'e']

# 使用 map + filter + zip
result = zip(
    map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)),
    letters
)
print(list(result))
# 输出：[(4, 'a'), (16, 'b'), (36, 'c'), (64, 'd'), (100, 'e')]

# 等价的生成器表达式（更 Pythonic）
result2 = zip((x**2 for x in numbers if x % 2 == 0), letters)
print(list(result2))
```

## 4. sorted(iterable, key=None, reverse=False)

![本地图片](./images/555.PNG)

```python
#基本排序
nums = [3, 1, 4, 1, 5, 9]
a=sorted(nums)
b=sorted(nums, reverse=True)
print(a)
print(b)
```

```python
# 使用 key 函数（类似 map 的转换）
words = ['apple', 'banana', 'cherry', 'date']
# 按字符串长度排序
a = sorted(words, key=len)
print(a)

# 按最后一个字母排序
b = sorted(words, key=lambda x: x[-1])
print(b)  # 分别以 a, e, e, y 结尾

# 对字典列表按某个键排序
students = [{'name': 'Alice', 'grade': 85}, {'name': 'Bob', 'grade': 92}, {'name': 'Charlie', 'grade': 78}]
c = sorted(students, key=lambda s: s['grade'], reverse=True)
print(c)
```

```python
# 对元组排序
tup = (3, 1, 4, 1, 5)
a=sorted(tup)
print(a)

# 对字符串排序（按 Unicode 码点）
b=sorted('hello')
print(b)
c=''.join(sorted('hello'))   # 得到排序后的字符串
print(c)
```

### 与 map 的关联
* key 函数的作用类似 map：它对每个元素应用一个转换，然后排序基于转换后的值。
* 与 map 的区别：sorted 不会保留原始元素的顺序，而是根据 key 返回值重新排列。
## 注意
* sorted 返回新列表，原可迭代对象不变。
* key 函数只会在每个元素上调用一次，然后缓存结果，所以即使处理大数据也高效。
* 如果需要就地排序（不生成新列表），使用列表的 .sort() 方法。

## 5.reversed(seq)

![本地图片](./images/666.PNG)

```python
# 基本用法
lst = [1, 2, 3, 4, 5]
rev = reversed(lst)
print(rev)
print(list(rev))

# 字符串反转
s = "hello"
s1 = ''.join(reversed(s))
print(s1)

# 元组反转
tup = (10, 20, 30)
tup1 = tuple(reversed(tup))
print(tup1)
```

```python
nums = [1, 2, 3, 4]
# 先映射再反转
squares = map(lambda x: x**2, nums)
a = list(reversed(list(squares)))   # 注意：map 对象只能遍历一次，需要先转为列表
print(a)

# 更好的写法：使用列表推导后反转
squares_rev = [x**2 for x in reversed(nums)]
print(squares_rev)
```

## 6.enumerate(iterable, start=0)

![本地图片](./images/777.PNG)

```python
# 基本用法
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits):
     print(i, fruit)

# 自定义起始索引
for i, fruit in enumerate(fruits, start=1):
     print(i, fruit)
```

```python
# 与map结合使用
fruits = ['apple', 'banana', 'cherry']
# 使用 map 和 enumerate 生成带索引的元组列表
indexed = list(map(lambda pair: pair, enumerate(fruits)))
print(indexed)

# 更常用的方式是直接用 enumerate 本身，无需 map
a=list(enumerate(fruits))
print(a)

# 使用 map 将索引和元素组合成字符串
result = list(map(lambda pair: f"{pair[0]}: {pair[1]}", enumerate(fruits)))
print(result)

# 使用列表推导更清晰
a=[f"{i}: {fruit}" for i, fruit in enumerate(fruits)]
print(a)
```

```python
# 与 sorted 结合：按索引排序或按值排序后保留原索引
scores = [85, 92, 78, 90]
# 按分数降序排序，同时保留原始索引
indexed_scores = list(enumerate(scores))  # [(0,85), (1,92), (2,78), (3,90)]
sorted_by_score = sorted(indexed_scores, key=lambda x: x[1], reverse=True)
print(sorted_by_score)

# 提取排序后的索引和分数
indices, sorted_scores = zip(*sorted_by_score)
print(indices)
print(sorted_scores)
```

### 注意
* enumerate 返回的迭代器每步产生一个元组，解包时常用 for i, val in enumerate(...):。
* 它与 map 的区别：enumerate 专门用于生成索引-值对，而 map 用于任意转换。
* 在需要同时使用索引和元素时，优先使用 enumerate，而不是 range(len(iterable))，后者更冗长且效率略低。
