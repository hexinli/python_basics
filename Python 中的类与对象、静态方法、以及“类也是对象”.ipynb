## 1、类与对象的基本用法
说明：类 Student 定义了实例的通用结构，每个实例拥有自己的 name 和 age，但共享类属性 school。

```python
class Student:
    """学生类，包含姓名和年龄"""

    # 类属性（所有实例共享）
    school = "第一中学"

    # 构造方法，初始化实例属性
    def __init__(self, name, age):
        self.name = name    # 实例属性
        self.age = age

    # 实例方法
    def introduce(self):
        print(f"我叫{self.name}，今年{self.age}岁，来自{self.school}")

    # 修改年龄的方法
    def have_birthday(self):
        self.age += 1
        print(f"{self.name} 长大了一岁，现在{self.age}岁")

# 创建对象（实例化）
stu1 = Student("小明", 15)
stu2 = Student("小红", 14)

stu1.introduce()   # 输出：我叫小明，今年15岁，来自第一中学
stu2.introduce()   # 输出：我叫小红，今年14岁，来自第一中学

stu1.have_birthday()   # 小明 长大了一岁，现在16岁
```

## 2. 静态方法（@staticmethod）
Python 类方法和静态方法的核心区别在于‌装饰器类型‌、‌第一个参数‌及‌访问权限‌。类方法使用 @classmethod 且首参数为 cls，可访问类属性；静态方法使用 @staticmethod 且无隐式参数，无法直接访问类或实例属性。

###### 类方法‌：必须使用 @classmethod 装饰器，第一个参数固定为 cls，代表类本身 。‌

###### 静态方法‌：必须使用 @staticmethod 装饰器，不需要 self 或 cls 参数，本质是类命名空间中的普通函数 。

```python
class MathUtils:
    """数学工具类，只包含静态方法"""
    my_num=10
    @staticmethod
    def add(x, y):
        return x + y

    @staticmethod
    def is_even(number):
        return number % 2 == 0

    @staticmethod
    def is_even_1():
        return MathUtils.my_num % 2 == 0

# 通过类名直接调用（推荐）
print(MathUtils.add(3, 5))        # 8
print(MathUtils.is_even(10))      # True
print(MathUtils.is_even_1())
# 也可以通过实例调用（不推荐，容易误导）
mu = MathUtils()
print(mu.add(4, 6))               # 10（依然有效，但没有意义）
```

```python
class Demo:
    class_var = 100

    def instance_method(self):
        return "实例方法"

    @classmethod
    def class_method(cls):
        return f"类方法，访问类变量 {cls.class_var}"

    @staticmethod
    def static_method():
        return "静态方法，无法访问类变量"

print(Demo.class_method())   # 类方法，访问类变量 100
print(Demo.static_method())  # 静态方法，无法直接访问类变量，如果有类变量需通过类名.变量的形式访问
```

## 3. 自定义的类属于“类的对象”（Python 中一切皆对象）
在 Python 中，类本身也是对象。当你定义一个类时，Python 会在内存中创建一个 type 类型的对象，这个对象（即类）具有属性和方法。因此，你可以：
###### 将类赋值给一个变量
###### 将类作为参数传递给函数
###### 在运行时动态为类添加属性或方法
###### 通过 type() 查看类的类型

```python
class Car:
    """简单的汽车类"""
    wheels = 4

    def __init__(self, brand):
        self.brand = brand

# 1. 类本身是 type 的实例
print(type(Car))      # <class 'type'>
print(isinstance(Car, type))  # True

# 2. 将类赋值给变量
MyClass = Car
my_car = MyClass("特斯拉")
print(my_car.brand)   # 特斯拉

# 3. 动态添加类属性
Car.color = "红色"
print(Car.color)      # 红色

# 4. 动态添加类方法（使用函数）
def honk(cls):
    return f"{cls.__name__} 按喇叭！"
Car.honk = classmethod(honk)
print(Car.honk())     # Car 按喇叭！

# 5. 类作为函数参数
def describe_class(cls):
    print(f"类名：{cls.__name__}")
    print(f"属性：{dir(cls)}")

describe_class(Car)
```

更深入的例子：类工厂（动态创建类）
因为类也是对象，你可以用 type() 动态创建新类，而不是用 class 关键字：

```python
# 使用 type(name, bases, dict) 动态创建类
Dog = type("Dog", (), {"sound": "汪汪", "speak": lambda self: self.sound})

d = Dog()
print(d.speak())   # 汪汪

# 甚至可以继承现有类
class Animal:
    def eat(self):
        return "吃东西"

Cat = type("Cat", (Animal,), {"sound": "喵喵"})
c = Cat()
print(c.eat())     # 吃东西（继承自Animal）
print(c.sound)     # 喵喵
```

### 4. 综合示例：将上述概念组合使用

```python
class TemperatureConverter:
    """温度转换器，提供静态工具方法"""

    @staticmethod
    def celsius_to_fahrenheit(c):
        return c * 9/5 + 32

    @staticmethod
    def fahrenheit_to_celsius(f):
        return (f - 32) * 5/9

class WeatherStation:
    """气象站类，记录某个城市的温度（摄氏度）"""

    # 类属性：记录所有创建的气象站实例
    all_stations = []

    def __init__(self, city, celsius):
        self.city = city
        self.celsius = celsius
        WeatherStation.all_stations.append(self)

    def report_celsius(self):
        print(f"{self.city} 当前温度：{self.celsius}°C")

    def report_fahrenheit(self):
        # 使用静态方法进行转换
        f = TemperatureConverter.celsius_to_fahrenheit(self.celsius)
        print(f"{self.city} 当前温度：{f:.1f}°F")

    @classmethod
    def count_stations(cls):
        return len(cls.all_stations)

    @staticmethod
    def is_valid_temperature(temp):
        """简单校验温度是否在合理范围（静态工具方法）"""
        return -100 <= temp <= 100

# 1. 创建实例
station1 = WeatherStation("北京", 25)
station2 = WeatherStation("上海", 30)

# 2. 调用实例方法
station1.report_celsius()    # 北京 当前温度：25°C
station1.report_fahrenheit() # 北京 当前温度：77.0°F

# 3. 使用静态方法（不需要实例）
print(TemperatureConverter.celsius_to_fahrenheit(0))  # 32.0
print(WeatherStation.is_valid_temperature(45))        # True

# 4. 使用类方法
print(f"当前有 {WeatherStation.count_stations()} 个气象站")  # 2

# 5. 演示“类也是对象”：动态添加一个类方法用于重置所有气象站列表
def reset_stations(cls):
    cls.all_stations = []
WeatherStation.reset = classmethod(reset_stations)

WeatherStation.reset()
print(f"重置后气象站数量：{WeatherStation.count_stations()}")  # 0

# 6. 甚至可以将类本身传递给函数
def show_class_info(cls):
    print(f"类名：{cls.__name__}, 文档：{cls.__doc__}")

show_class_info(WeatherStation)
```
